## Handling Outliers

```python
sns.boxplot(teen_m_c.age)
```

```python
teen_m_c.age.describe()
```

```python
q1 = teen_m_c.age.quantile(.25)
q3 = teen_m_c.age.quantile(.75)
```

```python
iqr = q3-q1
```

```python
(q1-1.5*iqr)
```

```python
(q3+1.5*iqr)
```

```python
teen_m_c=teen_m_c[(teen_m_c.age > (q1-1.5*iqr)) & (teen_m_c.age < (q3+1.5*iqr))]
```

## Elbow Method

```python
from sklearn.cluster import KMeans
inertia = []
for i in range(1,20):
    kmeans=KMeans(n_clusters=i,init = 'k-means++',n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1,20), inertia)
```

From the analysis of the above curve, I am selecting number of cluster as 4
